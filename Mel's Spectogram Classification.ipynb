### Short Time Fourier Transform Spectogram

```python
from scipy.signal import spectrogram
def plot_spectrogram(data, fs=1.0, nperseg=256, noverlap=128, nfft=256, cmap='magma'):
    f, t, Sxx = spectrogram(data, fs=fs, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
    plt.pcolormesh(t, f, 10 * np.log10(Sxx), shading='gouraud', cmap=cmap)
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
    plt.colorbar(label='Intensity [dB]')
    plt.title('Spectrogram')

# Plot the spectrogram of the first segment
# Sample number
i = 50
data_both_good = np.asarray(all_good_segments[i]['rel_time'], all_good_segments[i]['Torque'])
data_torque_good = np.asarray(all_good_segments[i]['Torque'])
plot_spectrogram(data_both_good, fs=1.0, nperseg=256, noverlap=128, nfft=256, cmap='magma')
plt.show()
plot_spectrogram(data_torque_good, fs=1.0, nperseg=256, noverlap=128, nfft=256, cmap='magma')
plt.show()

data_both_bad = np.asarray(all_bad_segments[i]['rel_time'], all_bad_segments[i]['Torque'])
data_torque_bad = np.asarray(all_bad_segments[i]['Torque'])
plot_spectrogram(data_both_bad, fs=1.0, nperseg=256, noverlap=128, nfft=256, cmap='magma')
plt.show()
plot_spectrogram(data_torque_bad, fs=1.0, nperseg=256, noverlap=128, nfft=256, cmap='magma')
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle

print('Good Mel features:', len(good_mels_features))
print('Bad Mel features:', len(bad_mels_features))

# Create labels for the features
good_labels = [1] * len(good_mels_features)  # Label 1 for good torque
bad_labels = [0] * len(bad_mels_features)    # Label 0 for bad torque

# Combine features and labels
features = good_mels_features + bad_mels_features
labels = good_labels + bad_labels

# Convert to numpy arrays
features = np.array(features)
labels = np.array(labels)

# Reshape features to (n_samples, n_features)
n_samples, n_mels, n_frames = features.shape
print("Data Shape:", n_samples, n_mels, n_frames)
features = features.reshape((n_samples, n_mels * n_frames))

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.4, random_state=42)

# Initialize the SVM classifier
svm_classifier = SVC(kernel='linear', random_state=42)

# Train the SVM classifier
svm_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print the results
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)


# Saving classifier using pickle 
pickle.dump(svm_classifier, open("svm_classiferv1", 'wb')) 
```

## Convolutional Neural Network - ResNet50 (Current Model Overfitting to Data)

```python
% script true
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision.models import resnet50
# Create labels for the features
good_labels = [1] * len(good_mels_features)  # Label 1 for good torque
bad_labels = [0] * len(bad_mels_features)    # Label 0 for bad torque

# Combine features and labels
features = np.array(good_mels_features + bad_mels_features)
labels = np.array(good_labels + bad_labels)

# Custom Dataset class
class MelSpectrogramDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        feature = self.features[idx]
        label = self.labels[idx]
        return torch.tensor(feature, dtype=torch.float32), torch.tensor(label, dtype=torch.long)

# Create dataset and dataloaders
dataset = MelSpectrogramDataset(features, labels)
train_size = int(0.4 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Modify ResNet-50
class ResNet50MelSpectrogram(nn.Module):
    def __init__(self, num_classes=2):
        super(ResNet50MelSpectrogram, self).__init__()
        self.resnet = resnet50(pretrained=True)
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        self.resnet.fc = nn.Linear(self.resnet.fc.in_features, num_classes)

    def forward(self, x):
        x = self.resnet(x)
        return x

# Instantiate model, loss function, and optimizer
model = ResNet50MelSpectrogram(num_classes=2)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 10
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.unsqueeze(1).to(device), labels.to(device)  # Add channel dimension

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / train_size
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}')

# Evaluation
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.unsqueeze(1).to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f'Test Accuracy: {accuracy:.2f}%')
```

## Simple Convolutional Neural Network

```python
% script true
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
import numpy as np

# Create labels for the features
good_labels = [1] * len(good_mels_features)  # Label 1 for good torque
bad_labels = [0] * len(bad_mels_features)    # Label 0 for bad torque

# Combine features and labels
features = np.array(good_mels_features + bad_mels_features)
labels = np.array(good_labels + bad_labels)

# Custom Dataset class
class MelSpectrogramDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        feature = self.features[idx]
        label = self.labels[idx]
        return torch.tensor(feature, dtype=torch.float32), torch.tensor(label, dtype=torch.long)

# Create dataset and dataloaders
dataset = MelSpectrogramDataset(features, labels)
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Simple CNN Model
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=2):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.fc1 = nn.Linear(64 * 64 * 13, 128)  # Adjust based on input size
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64 * 65 * 14)  # Flatten the tensor, updated for (129, 27) input size
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Instantiate model, loss function, and optimizer
model = SimpleCNN(num_classes=2)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 10
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.unsqueeze(1).to(device), labels.to(device)  # Add channel dimension

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / train_size
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}')

# Evaluation
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.unsqueeze(1).to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f'Test Accuracy: {accuracy:.2f}%')
```

## CNN Activation Map Visualization

```python
% script true
import torch
import torch.nn.functional as F
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Ensure the features array is of type float32
features = np.array(features, dtype=np.float32)


# Define a function to apply Grad-CAM
def grad_cam(model, input_tensor, target_class):
    # Hook the gradients of the target layer
    CAM_gradients = []
    CAM_activations = []

    def backward_hook(module, grad_in, grad_out):
        CAM_gradients.append(grad_out[0])
    
    def forward_hook(module, input, output):
        CAM_activations.append(output)
    
    # Register hook to the final convolutional layer
    final_conv = model.resnet.layer4[2].conv3
    handle1 = final_conv.register_forward_hook(forward_hook)
    handle2 = final_conv.register_backward_hook(backward_hook)

    # Forward pass
    model.eval()
    output = model(input_tensor)
    model.zero_grad()
    
    # Target for backprop
    target = output[0, target_class]
    
    # Backward pass
    target.backward()
    
    # Get the gradients and activations from the hooked layer
    gradients_tensor = CAM_gradients[0]
    activations_tensor = CAM_activations[0]
    
    # Pool the gradients across the width and height dimensions
    pooled_gradients = torch.mean(gradients_tensor, dim=[0, 2, 3])
    
    # Weight the channels by corresponding gradients
    for i in range(pooled_gradients.size(0)):
        activations_tensor[:, i, :, :] *= pooled_gradients[i]
    
    # Average the channels to get the heatmap
    heatmap = torch.mean(activations_tensor, dim=1).squeeze().cpu().numpy()
    
    # Normalize the heatmap
    heatmap = np.maximum(heatmap, 0)
    heatmap /= np.max(heatmap)
    
    # Remove the hooks
    handle1.remove()
    handle2.remove()
    
    return heatmap.detach().numpy()

# Function to visualize the heatmap on the spectrogram
def show_cam_on_image(img, mask):
    heatmap = cv2.applyColorMap(np.uint8(255 * mask), cv2.COLORMAP_JET)
    heatmap = np.float32(heatmap) / 255
    cam = heatmap + np.float32(img)
    cam = cam / np.max(cam)
    return np.uint8(255 * cam)

# Transform function for the input tensor
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485], std=[0.229])
])

# Example of a spectrogram input (replace with your actual data)
example_spectrogram = features[0].astype(np.float32)

# Prepare the input tensor (ensure it's 4D: [batch_size, channels, height, width])
input_tensor = transform(example_spectrogram).unsqueeze(0).to(device)

# Get the class activation map
target_class = 1  # Example target class (change as needed)
heatmap = grad_cam(model, input_tensor, target_class)

# Visualize the heatmap
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(example_spectrogram, aspect='auto', cmap='viridis')
plt.title('Original Mel Spectrogram')

plt.subplot(1, 2, 2)
cam_image = show_cam_on_image(example_spectrogram, heatmap)
plt.imshow(cam_image, aspect='auto')
plt.title('Class Activation Map')
plt.show()
```
